```python
import matplotlib.pyplot as plt
from palettable.colorbrewer.qualitative import Pastel1_9

#create data
names = ['GroupA', 'GeoupB', 'GroupC', 'GroupD']
size = [12,11,3,30]

#Create a circle at the centre of the plot
my_circle = plt.Circle((0,0), 0.7, color='white') # Corrected syntax for Circle constructor
plt.gca().add_artist(my_circle)

#create pie chart
plt.pie(size, labels=names, colors=Pastel1_9.hex_colors)
p = plt.gcf()
p.gca().add_artist(my_circle) # This line is redundant if my_circle is already added
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import squarify

#DataFrame
df = pd.DataFrame({'nb_people':[6,5,4,2], 'group':["GroupA","GroupB","GroupC","GroupD"]})

#Plot
squarify.plot(sizes=df['nb_people'], label=df['group'], alpha=.8)
plt.axis('off')
plt.show()
```

```python
import pandas as pd

data = pd.DataFrame({
    'lon':[-58, 2, 145, 30.32, -4.03, -73.57, 36.82, -38.5],
    'lat':[-34, 49, -41, 59.93, 5.33, 45.52, -1.29, -12.97],
    'name':['Buenos Aires', 'Paris', 'melbourne', 'St Petersbourg', 'Abidjan', 'Montreal', 'Nairobi', 'Salvador'],
    'value':[10,12,30,70,23,43,100,43]
}, dtype=str)
data
```

```python
import folium
from folium.plugins import MarkerCluster

m = folium.Map(location=[20,0], tiles="OpenStreetMap", zoom_start=2)
#Show Map
m
```

```python
for i in range(0,len(data)):
    folium.Marker(location=[data.iloc[i]['lat'], data.iloc[i]['lon']],
    popup=data.iloc[i]['name']).add_to(m)

#Show map
m
```
